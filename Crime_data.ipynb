```python
import pandas as pd

df = pd.read_csv('crime_csv_all_years.csv')

df.head()
```

```python
df.shape
```

```python
df.head()
```

```python
df = df[df.NEIGHBOURHOOD.notnull()]
```

```python
df = df.reset_index(drop=True)
```

```python
df = df.drop(columns=['HUNDRED_BLOCK', 'X','Y'])
```

```python
df.head()
```

```python
ctype = df['TYPE'].to_list()
cyear = df['YEAR'].to_list()
cmonth = df['MONTH'].to_list()
carea = df['NEIGHBOURHOOD'].to_list()
```

```python
crime_data = {'ctype':ctype,'cyear':cyear,'cmonth':cmonth,'carea':carea}
crime_data = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in crime_data.items() ]))
crime_data.head()
```

```python
crime_data.to_json('data.json',orient='records')
```

```python
crime_data.groupby('carea').size()
```

```python
df_e = pd.read_csv('employment.csv')
df_e
```

```python
df_e = df_e.set_index('id').transpose()
```

```python
# df_e = df_e.reset_index()
```

```python
df_e.head()
```

```python
carea = df_e['carea'].to_list()
emp = df_e['Employment rate'].to_list()
unemp = df_e['Unemployment rate'].to_list()
```

```python
emp_data = {'carea':carea,'emp':emp,'unemp':unemp}
emp_data = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in emp_data.items() ]))
emp_data.head()
```

```python
emp_data.to_json('empdata.json',orient='records')
```

```python
emp_data.carea.unique()
```

```python
crime_data.head()
```

```python
emp_data.carea = emp_data.carea.str.strip()
joined_data = crime_data.merge(emp_data,how = "inner", suffixes = ('_x', '_y'), on = ['carea'])
```

```python
ctype = joined_data['ctype'].to_list()
cyear = joined_data['cyear'].to_list()
cmonth = joined_data['cmonth'].to_list()
carea = joined_data['carea'].to_list()
emp = joined_data['emp'].to_list()
unemp = joined_data['unemp'].to_list()
```

```python
data = {'Offense':ctype,'YEAR':cyear,'MONTH':cmonth,'NEIGHBOURHOOD':carea,'emp':emp,'unemp':unemp}
data = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in data.items() ]))
data.head()
```

```python
data.to_json('cdata.json',orient='records')
```
